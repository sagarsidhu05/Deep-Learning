#Question 3: Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.
#Answer:
Purpose of Activation Function in Neural Networks

An activation function introduces non-linearity into a neural network.
Without it, every layer would perform only a linear transformation, and stacking multiple layers would still behave like a single linear model.
By adding non-linear behavior, activation functions allow the network to learn complex relationships, capture intricate patterns, and approximate any continuous function.
They also control the range of neuron outputs and influence how gradients flow during backpropagation, affecting training speed and stability.

Sigmoid

The sigmoid function maps input values to a range between 0 and 1 using the formula

Ïƒ(x)=1/1+eâˆ’x

It is smooth and differentiable, making it useful for tasks that require probabilities or binary classification.
However, for very large positive or negative inputs, the gradient becomes extremely small, leading to the vanishing gradient problem and slower learning.

Tanh

The hyperbolic tangent (tanh) function maps inputs to the range â€“1 to 1,

tanh(x)=ex-e-x/ex+e-x
â¡

Because its output is zero-centered, it often leads to faster convergence than sigmoid.
Yet it can still suffer from saturation, where extreme input values produce very small gradients and slow training.

ReLU

The Rectified Linear Unit (ReLU) is defined as

f(x)=max(0,x).

It is simple, computationally efficient, and helps alleviate the vanishing gradient issue because positive inputs have a constant gradient of 1.
ReLU has become the default activation in many deep learning models.
Its main drawback is the â€œdying ReLUâ€ problem, where neurons can become inactive (outputting 0 for all inputs) if the weights update in an unfavorable way.

Summary

Activation functions are essential for enabling neural networks to learn complex, non-linear mappings.

Sigmoid is good for probabilistic outputs but suffers from vanishing gradients.

Tanh improves on sigmoid by being zero-centered but can still saturate.

ReLU is computationally simple and widely used, though it risks inactive neurons.   

#Question 4: What is the difference between Loss function and Cost function in neural networks? Provide examples.
#Answer:
Difference Between Loss Function and Cost Function in Neural Networks:

1.Loss Function

.The loss function measures the error for a single training example.

.It quantifies how far the modelâ€™s prediction is from the true target for that one sample.

.During training, the model computes the loss for each data point to guide weight updates.

Examples:

.Mean Squared Error (MSE) loss for one sample

.Binary Cross-Entropy loss for a single input-output pair

.Hinge loss for a single classification instance

2.Cost Function

.The cost function represents the average of the loss values across the entire training dataset or a batch.

.It summarizes overall model performance and is the value that the optimization algorithm actually minimizes.

Examples:

.Average Mean Squared Error over all training samples

.Average Categorical Cross-Entropy over a mini-batch or full dataset

Key Idea:

.Loss function: Error for one sample.

.Cost function: Aggregate (mean or sum) of losses over many samples.

In practice, people sometimes use the terms interchangeably, but strictly speaking, the loss is per example, while the cost is the overall measure minimized during training.

#Question 5: What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.
#Answer:
Role of Optimizers in Neural Networks:

Optimizers are algorithms that adjust the weights and biases of a neural network to minimize the cost function.
During training, the network computes the gradient of the loss with respect to each parameter.
The optimizer then uses these gradients to update the parameters so that the model gradually learns patterns in the data.

Key responsibilities of an optimizer:

.Determine the direction in which to move parameters (using gradients).

.Decide the size of the step (learning rate or adaptive step sizes).

.Balance convergence speed with stability to avoid overshooting or getting stuck in local minima.

1.Gradient Descent

.Idea: Moves parameters in the negative direction of the gradient of the cost function.

.Types: Batch, Stochastic, and Mini-batch Gradient Descent.

.Strengths: Simple and effective for convex problems.

.Limitations: Fixed learning rate; can be slow to converge and may get trapped in local minima or saddle points.

2.RMSprop

.Concept: Root Mean Square Propagation.

.How it Works: Maintains a moving average of squared gradients for each parameter and divides the learning rate by the root of this average.

.Advantages: Adjusts learning rates adaptively for each parameter, helping handle non-stationary objectives and speeding convergence.

.Use Cases: Effective for recurrent neural networks and problems with noisy gradients.

3.Adam

.Concept: Adaptive Moment Estimation.

.How it Works: Combines ideas from RMSprop and momentum. It keeps an exponential moving average of past gradients (momentum) and their squares, adapting the learning rate for each parameter.

.Advantages: Fast convergence, robust to noisy data, minimal tuning required.

.Use Cases: Widely used default optimizer for many deep learning tasks.

Summary:

Optimizers are crucial for efficient and stable training of neural networks.

.Gradient Descent is simple but can be slow and sensitive to learning rates.

.RMSprop adapts learning rates using squared gradient averages, improving performance on non-stationary or noisy problems.

.Adam combines momentum and adaptive learning rates, making it one of the most popular and versatile choices in modern deep learning.


#Question 6: Write a Python program to implement a single-layer perceptron from scratch using NumPy to solve the logical AND gate.


```python
# Single-layer Perceptron for AND Gate using NumPy, Matplotlib, and TensorFlow/Keras

import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers

# -----------------------------
# 1. Prepare the AND gate data
# -----------------------------
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]], dtype=float)
y = np.array([[0],
              [0],
              [0],
              [1]], dtype=float)   # AND gate labels

# -----------------------------
# 2. Build the Model
# -----------------------------
# A single Dense layer with 1 neuron and a sigmoid activation
model = keras.Sequential([
    layers.Dense(1, input_dim=2, activation='sigmoid')
])

model.compile(optimizer='sgd',           # Stochastic Gradient Descent
              loss='binary_crossentropy',
              metrics=['accuracy'])

# -----------------------------
# 3. Train the Model
# -----------------------------
history = model.fit(X, y,
                    epochs=200,
                    verbose=0)  # silent training

# -----------------------------
# 4. Evaluate the Model
# -----------------------------
print("\nPredictions after training:")
for xi in X:
    pred = model.predict(xi.reshape(1, 2), verbose=0)
    print(f"{xi} -> {int(pred > 0.5)} (probability: {pred[0][0]:.3f})")

# -----------------------------
# 5. Plot Training Loss
# -----------------------------
plt.figure(figsize=(6,4))
plt.plot(history.history['loss'])
plt.title('Training Loss for AND Gate Perceptron')
plt.xlabel('Epoch')
plt.ylabel('Binary Cross-Entropy Loss')
plt.grid(True)
plt.show()
```

#Question 7: Implement and visualize Sigmoid, ReLU, and Tanh activation functions using Matplotlib.

```python
# Activation Functions Visualization: Sigmoid, ReLU, Tanh
import numpy as np
import matplotlib.pyplot as plt

# 1. Define input range
x = np.linspace(-10, 10, 400)

# 2. Define activation functions
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def relu(z):
    return np.maximum(0, z)

def tanh(z):
    return np.tanh(z)

# 3. Compute outputs
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

# 4. Plot all three
plt.figure(figsize=(10, 6))

plt.plot(x, y_sigmoid, label="Sigmoid", color="blue")
plt.plot(x, y_relu,    label="ReLU",    color="red")
plt.plot(x, y_tanh,    label="Tanh",    color="green")

plt.title("Activation Functions: Sigmoid, ReLU, Tanh")
plt.xlabel("Input (x)")
plt.ylabel("Output f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

#Question 8: Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset. Print the training accuracy.


```python
# Multilayer Neural Network on MNIST using Keras
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers

# 1. Load and preprocess the MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normalize pixel values to [0, 1]
x_train = x_train.astype("float32") / 255.0
x_test  = x_test.astype("float32") / 255.0

# Flatten 28x28 images to 784-dimensional vectors
x_train = x_train.reshape(-1, 784)
x_test  = x_test.reshape(-1, 784)

# Convert labels to one-hot encoded vectors
y_train = keras.utils.to_categorical(y_train, 10)
y_test  = keras.utils.to_categorical(y_test, 10)

# 2. Build the neural network model
model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# 3. Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# 4. Train the model
history = model.fit(
    x_train,
    y_train,
    epochs=5,
    batch_size=128,
    verbose=1
)

# 5. Print final training accuracy
final_train_acc = history.history['accuracy'][-1]
print("\nFinal Training Accuracy: {:.4f}".format(final_train_acc))
```

#Question 9: Visualize the loss and accuracy curves for a neural network model trained on the Fashion MNIST dataset. Interpret the training behavior

```python
# Train a neural network on Fashion MNIST and visualize loss & accuracy curves
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers

# 1. Load and preprocess the Fashion MNIST data
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

x_train = x_train.astype("float32") / 255.0
x_test  = x_test.astype("float32") / 255.0
x_train = x_train.reshape(-1, 784)
x_test  = x_test.reshape(-1, 784)

y_train = keras.utils.to_categorical(y_train, 10)
y_test  = keras.utils.to_categorical(y_test, 10)

# 2. Build a simple neural network model
model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# 3. Train the model and store training history
history = model.fit(
    x_train,
    y_train,
    epochs=10,
    batch_size=128,
    validation_split=0.2,
    verbose=1
)

# 4. Plot training vs validation loss and accuracy
epochs = range(1, len(history.history['loss']) + 1)

plt.figure(figsize=(12,5))

# Loss plot
plt.subplot(1, 2, 1)
plt.plot(epochs, history.history['loss'], 'b-', label='Training Loss')
plt.plot(epochs, history.history['val_loss'], 'r-', label='Validation Loss')
plt.title('Loss Curves')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Accuracy plot
plt.subplot(1, 2, 2)
plt.plot(epochs, history.history['accuracy'], 'b-', label='Training Accuracy')
plt.plot(epochs, history.history['val_accuracy'], 'r-', label='Validation Accuracy')
plt.title('Accuracy Curves')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Fraud Detection Example using Keras
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, roc_auc_score
from tensorflow import keras
from tensorflow.keras import layers

# ---------------------------
# 1. Generate synthetic dataset
# ---------------------------
np.random.seed(42)
N = 10000  # total samples
n_features = 10

# Random features
X = np.random.randn(N, n_features)

# Labels: 1% fraud (imbalanced)
y = np.zeros(N)
y[:int(0.01 * N)] = 1
np.random.shuffle(y)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)

# Feature scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)

# ---------------------------
# 2. Compute class weights
# ---------------------------
from sklearn.utils import class_weight
class_weights = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train),
    y=y_train
)
class_weights_dict = {0: class_weights[0], 1: class_weights[1]}
print("Class weights:", class_weights_dict)

# ---------------------------
# 3. Build the neural network
# ---------------------------
model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(n_features,)),
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# ---------------------------
# 4. Train the model
# ---------------------------
history = model.fit(
    X_train,
    y_train,
    epochs=20,
    batch_size=128,
    class_weight=class_weights_dict,
    validation_split=0.2,
    verbose=1
)

# ---------------------------
# 5. Evaluate the model
# ---------------------------
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

print("\nClassification Report:")
print(classification_report(y_test, y_pred, digits=4))

auc_score = roc_auc_score(y_test, y_pred_prob)
print("ROC AUC:", auc_score)

# ---------------------------
# 6. Plot training curves
# ---------------------------
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

# Loss
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Curve')
plt.xlabel('Epoch')
plt.ylabel('Binary Crossentropy Loss')
plt.legend()
plt.grid(True)

# Accuracy
plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Curve')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
